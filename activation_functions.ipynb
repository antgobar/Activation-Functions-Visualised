```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
```

```python
%matplotlib inline
```

```python
# ReLU

def relu(inputs):
    return np.maximum(0,inputs)

def plot_relu(min_val, max_val, weight, bias):
    x = np.linspace(min_val, max_val ,100)
    y = relu(x * weight + bias)
    plt.plot(x, y)
    plt.xlabel("Inputs")
    plt.ylabel("Activation value")
    plt.title(f"ReLU activation with weight = {weight} and bias = {bias}")
    plt.grid()
    plt.show()    

plot_relu(min_val = -6,
             max_val = 6,
             weight = 1.5,
             bias = -3)
```

```python
# Sigmoid

def sigmoid(inputs):
    return 1/(1 + np.exp(-inputs))

def plot_sigmoid(min_val, max_val, weight, bias):
    x = np.linspace(min_val, max_val ,100)
    y = sigmoid(x * weight + bias)
    plt.xlabel("Inputs")
    plt.ylabel("Activation value")
    plt.title(f"Sigmoid activation with weight = {weight} and bias = {bias}")
    plt.grid()
    plt.plot(x,y)

plot_sigmoid(min_val = -6,
          max_val = 6,
          weight = 1.5,
          bias = -3)
```

## Softmax

$\large\ S(x_i) = \frac{e^{x_i}}{\sum_{} e^{x_i}}$

```python
# Softmax 

np.set_printoptions(suppress=True)
layer_outputs = np.array([[4.8, 1.21, 2.385],
                           [8.9, -1.81, 0.2],
                           [1.41, 1.051, 0.026]])

exp_values = np.exp(layer_outputs - np.max(layer_outputs, axis = 1, keepdims = True))
probabilities = exp_values / np.sum(exp_values, axis = 1, keepdims = True)
print(probabilities)
```
